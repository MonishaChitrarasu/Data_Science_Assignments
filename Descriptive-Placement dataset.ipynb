```python
des_data=pd.DataFrame(index=["mean","median","mode","null","nan","count","min","25%","50%","75%","100%","IQR","1.5rule","Lesser","Greater","std"],columns=quan)
for column in quan:
        des_data[column]["mean"]=dataset[column].mean()
        des_data[column]["median"]=dataset[column].median()
        des_data[column]["mode"]=dataset[column].mode()[0]
        des_data[column]["null"]=dataset[column].isnull().sum()
        des_data[column]["nan"]=dataset[column].isna().sum()
        des_data[column]["count"]=dataset[column].count()
        des_data[column]["min"]=dataset[column].min()
        des_data[column]["25%"]=np.percentile(dataset[column],25)
        des_data[column]["50%"]=np.percentile(dataset[column],50)
        des_data[column]["75%"]=np.percentile(dataset[column],75)
        des_data[column]["100%"]=np.percentile(dataset[column],100)
        des_data[column]["IQR"]=des_data[column]["75%"]-des_data[column]["25%"]
        des_data[column]["1.5rule"]=1.5*des_data[column]["IQR"]
        des_data[column]["Lesser"]=des_data[column]["25%"]-des_data[column]["1.5rule"]
        des_data[column]["Greater"]=des_data[column]["25%"]+des_data[column]["1.5rule"]
        des_data[column]["std"]=dataset[column].std()
        
        
```

```python
des_data
```

```python
lesser=[]
greater=[]
for i in quan:
    if(des_data[i]["Lesser"]>des_data[i]['min']):
        lesser.append(i)
    if(des_data[i]["Greater"]<des_data[i]['100%']):
        greater.append(i)
print("Lesser Range",lesser)
print("Greater Range",greater)
    
```

```python
for i in lesser:
    dataset[i][dataset[i]<des_data[i]['Lesser']]=des_data[i]['Lesser']
        #print(dataset[i])
for j in greater:
    dataset[j][dataset[j]>des_data[j]['Greater']]=des_data[j]['Greater']
```

```python
def changeoutlier(des_data,dataset):
    for i in lesser:
        dataset[i][dataset[i]<des_data[i]['Lesser']]=des_data[i]['Lesser']
        #print(dataset[i])
    for j in greater:
        dataset[j][dataset[j]>des_data[j]['Greater']]=des_data[j]['Greater']
        #print(dataset[j])
    return des_data
```

```python
changeoutlier(des_data,dataset)
```
